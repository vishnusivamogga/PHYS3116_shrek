```python
%pip install numpy pandas matplotlib plotly galpy astropy


# Imported a library for numerical operations in Python
import numpy as np
# Imported a library for data manipulation and analysis
import pandas as pd
# Imported a library for creating static, animated, and interactive visualizations in Python
import matplotlib.pyplot as plt




import plotly.express as px
from galpy.orbit import Orbit
from galpy.potential import McMillan17
from astropy import units as u
import pandas as pd
import astropy as ap
from astropy.coordinates import SkyCoord, ICRS, Galactic, Galactocentric
from galpy.potential import MWPotential2014
from galpy.orbit import Orbit




# Read HarrisPartI
Harris1 = pd.read_csv("data/HarrisPartI.csv")
# Read HarrisPartIII
Harris3 = pd.read_csv("data/HarrisPartIII.csv")

```

```python
# Merge Harris1 and Harris3 on the "ID" column
dataframe = pd.merge(Harris1, Harris3, on="ID")


# Display the keys (column names) of the merged DataFrame so i t's easier to see what data we have
print(dataframe.keys())
```

```python
# Extract the X, Y, Z coordinates and assign units of kiloparsecs (kpc)
x = dataframe["X"].values * u.kpc
y = dataframe["Y"].values * u.kpc  
```

```python
# Creating an empty figure and axis for the xy-plane plot
fig, xy = plt.subplots(1, figsize=(10, 5))


# Plot the coordinates in the xy-plane plot
xy.scatter(x, y)


# Plot the accreted candidates
xy.set_xlabel("x (kpc)", fontsize=15)
xy.set_ylabel("y (kpc)", fontsize=15)
xy.grid(True, linestyle='--', alpha=0.8)


# Plot a red line which represents the Galactic Center at x=8 kpc
xy.axvline(x=8, color='red', linestyle='--', linewidth=2, label='x=8')


# Plot a yellow point at (0,0) to represent the Sun's position)=
xy.scatter(0, 0, color='yellow', s=200, label='Point (0,0)')
```
